## Matplotlib example workflow

```python
# 1. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#2. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#3. Setup Plot
fig, ax = plt.subplots(figsize=(5,5)) #width, height

#4. Plot Data
ax.plot(x,y)

#5.Customize Plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel = "y-axis")

#6. Save and show (you save the whole figure)
fig.savefig("sample-plot.png")
```

```python
# Plot from dict
butter_prices = {"Almond" : 10,
                 "Peanut" : 8,
                 "Cashew" : 12}
fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title="Don's Butter Store", ylabel= "Price ($)");
```

```python
#horizontal bar graph
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()));
```

```python
#Histograms
x = np.random.randn(1000) #create a numbers for normal distribution
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Two options for subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#Plotting to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(butter_prices.keys(), butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
#Subplot option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#plot to each different index
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(butter_prices.keys(), butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### Which one should we use?? Pyplot vs matplotlib OO methods

* When plotting something quickly, we can use pyplot method
* When plotting something more advanced, use OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#Pyplot Method
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
## OO Method with pyplot
fig, ax = plt.subplots(figsize=(10,5))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
#ax.set_xlim([45,100]); #changes the x axis 
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#plotting data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

#customizing the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

#add a legend
ax.legend(*scatter.legend_elements(), title="Target")

#add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
# Subplot of chol, age,thalach

fig, (ax1, ax2) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10), sharex=True)

#add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

#customizing
ax1.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")
#legend 
ax1.legend(*scatter.legend_elements(), title="Target")

#Mean Line
ax1.axhline(over_50["chol"].mean(), linestyle='--');


#add data to ax2
scatter_2= ax2.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

#customize
ax2.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")
#legend
ax2.legend(*scatter_2.legend_elements(), title="Target")
ax2.axhline(over_50["thalach"].mean(), linestyle='--');

#Add a title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
